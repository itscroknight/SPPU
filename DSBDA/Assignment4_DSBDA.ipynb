Practicle 4: Create a Linear Regression model using python to predict home prise using Boston Housing dataset.

```python
import pandas as pd
```

Impoted Pandas Library.

```python
df = pd.read_csv("/content/BostonHousing.csv")
```

File Readed using read_csv function.

```python
df
```

```python
df.shape
```

```python
df.isnull().any()
```

Checkedd for null Values.

```python
x = df.drop('medv', axis=1)
y = df['medv']
```

Divided data into x and y to perform further operations.

In y we have house price that we will predict from linear regression.

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

Imported "train_test split" from sklearn.model_selection library.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0, test_size=0.25)
```

Now, defining x_train, x_test, y_train, y_test to split data and splitting it in train:test = 75%:25%

```python
x_train
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

Imported Linear Regression from sklearn.linear_model library.

```python
regressor = LinearRegression()
```

Defined object of linear regression as "regressor".

```python
regressor.fit(x_train, y_train)
```

Used regressor.fit to fit data of x_train and y_train.

```python
regressor.coef_
```

X coefficient value.

```python
regressor.intercept_
```

Y-intercept value in line equation y = mx + b.

```python
y_predict_test = regressor.predict(x_test)
```

Passing x_test to predict its value through linear regression.

```python
y_predict_test
```

```python
y_predict_train = regressor.predict(x_train)
```

```python
result = pd.DataFrame({'Actual': y_test, 'Predicted': y_predict_test})
```

Created a Dataframe to display 'Actual' and 'Predicted' values.

```python
result
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

Imported mean_absolute_percentage_error from sklearn.metrics library.

```python
mean_absolute_percentage_error(y_test, y_predict_test)
```

Calculated mean_absolute_percentage_error for y_test and predicted.

```python
from sklearn.metrics import r2_score
```

Imported R2_score from sklearn.metrics library.

```python
r2_score(y_test, y_predict_test)
```

r2 score checks the accuracy of the model which comes out to be 63.54% here.

```python
from matplotlib import pyplot as plt
plt.scatter(y_train, y_predict_train, c = 'blue', marker = 'o', label='Traning Data')
plt.scatter(y_test, y_predict_test, c = 'green', marker = 's', label='Test Data')
plt.xlabel('True Values')
plt.ylabel('Predicted')
plt.title("True Values vs Predicted Values")
plt.legend(loc='upper left')
plt.plot()
plt.show()
```

From matplotlib imported pyplot to plot the graph.

Plotted a graph for 'Actual vs predicted' values in test and train data.
