**Practical 5:**  Implement logistic regression using python to perform classifiaction on social network_ads , cv dataset.  




```python
import pandas as pd
import numpy as np
```

Imported Pandas and Numpy Libraries.

```python
df = pd.read_csv("/content/Social_Network_Ads_ASS 5_dataset.csv")
```

Using read_csv readed the dataset "Social_Network_Ads".

```python
df
```

```python
x = df[['Age', 'EstimatedSalary']]
y = df['Purchased']
```

Seperated the x and y. X contains Age and salary. Y contains Purchased column.

```python
x
```

```python
y
```

```python
x.isnull().any()
y.isnull().any()
```

Checked for null values.

```python
from sklearn import preprocessing
```

```python
min_max_scalar= preprocessing.MinMaxScaler()
x_scaled = min_max_scalar.fit_transform(x)
x_scaled
```

Preprocessed the data using MinMaxScalar Normalization method. So, that the values are in range.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, random_state = 0, test_size = 0.25)
```

Used train_test_split to split the data in train and test in ratio 75:25.

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

Imported Logistic Regression andd defined its object "classifier".

```python
import seaborn as sns
```

```python
sns.countplot(x=y)
```

Ploted a graph to check count of Purchased column's classes.

```python
y.value_counts()
```

```python
classifier.fit(x_train, y_train)
```

Using fit we fitted the values of x_train and y_train to train the model.

```python
y_predict = classifier.predict(x_test)
```

Predicted the y_predict using the x_test values.

```python
y_predict
```

```python
result = pd.DataFrame({'Actual': y_test, 'Predicted': y_predict})
```

Comparing the Actual an Predicted values.

```python
result
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Age')
plt.ylabel('Salary')
plt.scatter(x['Age'], x['EstimatedSalary'], c=y)
```

Plotted a graph to show variation in values and we can see it shows "s" curve i.e. Logistic Regression.

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_predict )
```

Using Confusion Matrix we plotted a matrix to show TP, TN, FP, FN values count that the model predicted.

```python
from sklearn.metrics import ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_predict, labels=classifier.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classifier.classes_)
disp.plot()
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict))
```

This classification_report shows the precision, recall, support and f1-score of the classified classes.
